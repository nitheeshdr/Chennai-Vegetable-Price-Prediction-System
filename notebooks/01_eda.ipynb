# Chennai Vegetable Price EDA
Explore raw data distributions, missing values, and price patterns.

```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_theme(style='darkgrid')
%matplotlib inline
```

```python
# Load raw mandi data
mandi = pd.read_parquet('../data/raw/mandi_raw.parquet')
print(f'Shape: {mandi.shape}')
mandi.head()
```

```python
# Vegetable distribution
top_veg = mandi['vegetable_name'].value_counts().head(20)
top_veg.plot(kind='barh', figsize=(10,6), title='Record count per vegetable')
plt.tight_layout()
plt.show()
```

```python
# Price time series for tomato
tomato = mandi[mandi['vegetable_name']=='tomato'].sort_values('date')
plt.figure(figsize=(14,4))
plt.plot(tomato['date'], tomato['modal_price'])
plt.title('Tomato Modal Price — Koyambedu')
plt.ylabel('Price (Rs/kg)')
plt.tight_layout()
plt.show()
```

```python
# Missing value heatmap
plt.figure(figsize=(10,5))
sns.heatmap(mandi[['modal_price','min_price','max_price','arrival_qty']].isna(), cbar=False)
plt.title('Missing values')
plt.show()
```

```python
# Seasonal price patterns
mandi['month'] = pd.to_datetime(mandi['date']).dt.month
seasonal = mandi.groupby(['vegetable_name','month'])['modal_price'].mean().unstack()
top5 = mandi['vegetable_name'].value_counts().head(5).index
seasonal.loc[top5].T.plot(figsize=(12,5), title='Monthly Average Price by Vegetable')
plt.ylabel('Price (Rs/kg)')
plt.tight_layout()
plt.show()
```
