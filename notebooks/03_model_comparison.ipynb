# Model Comparison
Load comparison report and visualize RMSE/MAE/MAPE across all models.

```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
%matplotlib inline
```

```python
df = pd.read_csv('../data/model_artifacts/model_comparison.csv')
print(df.shape)
df.head()
```

```python
# Average RMSE per model
avg = df.groupby('model')[['rmse','mae','mape']].mean().sort_values('rmse')
avg['rmse'].plot(kind='bar', figsize=(10,5), title='Average RMSE by Model (lower=better)', color='steelblue')
plt.ylabel('RMSE (Rs/kg)')
plt.tight_layout()
plt.show()
print(avg.round(2))
```

```python
# Per-vegetable best model heatmap
pivot = df.pivot_table(index='vegetable', columns='model', values='rmse')
plt.figure(figsize=(14,8))
sns.heatmap(pivot, annot=True, fmt='.1f', cmap='RdYlGn_r', linewidths=0.5)
plt.title('RMSE per vegetable per model')
plt.tight_layout()
plt.show()
```
